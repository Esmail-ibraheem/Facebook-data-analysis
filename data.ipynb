```python
import pandas as pd
import numpy as np

# Simulate user data
users = pd.DataFrame({
    'id': range(100),
    'name': [f'User {i}' for i in range(100)],
    'age': np.random.randint(18, 65, 100),
    'location': np.random.choice(['Sanaa', 'Aden', 'Alhodida', 'Taiz'], 100),
    'interests': [np.random.choice(['Sports', 'Music', 'Movies', 'Books'], 3) for _ in range(100)]
})

# Simulate user interactions (e.g., likes, comments, shares)
interactions = pd.DataFrame({
    'id': np.random.choice(users['id'], 1000),
    'post_id': np.random.randint(1, 100, 1000),
    'interaction_type': np.random.choice(['like', 'comment', 'share'], 1000)
})

# Merge user data with interaction data
data = pd.merge(users, interactions, on='id')

# Save data to a CSV file
data.to_csv('facebook_data.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('facebook_data.csv')

# Analyze the data
print("Top 5 locations by user count:")
print(data['location'].value_counts().head(5))

print("\nTop 5 interests by user count:")
print(data['interests'].value_counts().head(5))

# Plot the distribution of user ages
plt.hist(data['age'], bins=10)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of User Ages')
plt.show()

# Plot the distribution of user interactions by type
plt.bar(data['interaction_type'].value_counts().index, data['interaction_type'].value_counts())
plt.xlabel('Interaction Type')
plt.ylabel('Frequency')
plt.title('Distribution of User Interactions')
plt.show()
```
